```
words = ['this','is', 'the','best','mac','ever']
wordsRdd = sc.parallelize(words)
```

```
wordsRdd.reduce(lambda x,y: x if len(x)<len(y) else y)
```

```
def mod(x,y):
    if len(x)<len(y):
        return str(x).upper()
    else:
        return str(y).lower()
```

```
wordsRdd.reduce(mod)
```

```
wordsRdd.getNumPartitions()
```

```
wordsRdd = wordsRdd.repartition(2)
wordsRdd.getNumPartitions()
```
